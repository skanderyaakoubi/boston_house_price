```python
boston = load_boston()

# Convertir en DataFrame pandas
df = pd.DataFrame(data=boston['data'], columns=boston['feature_names'])

# Ajouter la colonne cible (target)
df['target'] = boston['target']
```

```python
import seaborn as sns
corr_matrix = df.corr()

# Crer une heatmap de la matrice de corrlation
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Matrice de Corrlation')
plt.show()
```

```python
plt.figure(figsize=(15, 10))  # Taille de la figure

for i, column in enumerate(df.columns[:-1]):  # Exclure la colonne 'target' elle-mme
    plt.subplot(3, 4, i+1)  # Crer des sous-graphes
    plt.scatter(df[column], df['target'], alpha=0.5)
    plt.title(f'{column} vs target')
    plt.xlabel(column)
    plt.ylabel('target')
```

```python
y=df.target
x=df.drop(columns="target",axis=0)
```

```python
x=df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=12)
```

```python
x_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
type(x_train)
```

```python
x_test = scaler.fit_transform(x_test)
```

Model ttraining

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
model = LinearRegression()

# Entraner le modle
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)

# valuer le modle
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
from sklearn.linear_model import Ridge
ridge_model = Ridge(alpha=1.0)
ridge_model.fit(x_train, y_train)

# Faire des prdictions
y_pred_ridge = ridge_model.predict(x_test)

# valuer le modle
print("MSE:", mean_squared_error(y_test, y_pred_ridge))
print("R:", r2_score(y_test, y_pred_ridge))
```

```python
from sklearn.linear_model import Lasso

# Crer et entraner le modle Lasso
lasso_model = Lasso(alpha=0.1)
lasso_model.fit(x_train, y_train)

# Faire des prdictions
y_pred_lasso = lasso_model.predict(x_test)

# valuer le modle
print("MSE:", mean_squared_error(y_test, y_pred_lasso))
print("R:", r2_score(y_test, y_pred_lasso))
```

```python
from sklearn.ensemble import RandomForestRegressor

# Crer et entraner le modle de fort alatoire
rf_model = RandomForestRegressor(n_estimators=100)
rf_model.fit(x_train, y_train)

# Faire des prdictions
y_pred_rf = rf_model.predict(x_test)

# valuer le modle
print("MSE:", mean_squared_error(y_test, y_pred_rf))
print("R:", r2_score(y_test, y_pred_rf))
```

```python
from sklearn.svm import SVR

# Crer et entraner le modle SVR
svr_model = SVR(kernel='linear')
svr_model.fit(x_train, y_train)

# Faire des prdictions
y_pred_svr = svr_model.predict(x_test)

# valuer le modle
print("MSE:", mean_squared_error(y_test, y_pred_svr))
print("R:", r2_score(y_test, y_pred_svr))
```
